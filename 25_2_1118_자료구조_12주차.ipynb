# 순차 탐색 알고리즘

```python
# 순차 탐색 알고리즘 함수
def sequential_search(A, key, low, high):
  for i in range(low, high + 1):
    if A[i] == key:
      return i
  return -1
```

```python
array = [3, 2, 1, 5, 4, 7, 6, 9, 8, 0]

n = len(array)
print(sequential_search(array, 7, 0, n - 1))
```

# 이진 탐색 알고리즘

```python
# 이진 탐색 알고리즘 함수   # 정렬 되어 있어야 함
def binary_search(A, key, low, high):

  # 탐색 실패
  if (low > high):
    return -1

  middle = (low + high) // 2

  # 탐색 성공
  if (key == A[middle]):
    return middle

  # 키가 미들보다 작음
  elif (key < A[middle]):
    return binary_search(A, key, low, middle - 1)

  # 키가 미들보다 큼
  else:
    return binary_search(A, key, middle + 1, high)
```

```python
array = [1, 2, 4, 8, 16, 32, 64, 128]

n = len(array)
print(binary_search(array, 32, 0, n - 1))
```

# 문자열을 위한 해시함수

```python
ord('c') # 문자 -> 아스키
```

```python
def hashFn(key, M):   # M == 테이블 크기
  sum = 0
  for c in key:
    sum = sum + ord(c)
  return sum % M
```

```python
hashFn('Mika', 12)
```

# 선형 조사법

```python
M = 13
table = [None] * M

def hashFn(key):
  return key % M

# 삽입 연산
def insert(key):
  i = hashFn(key)     # 해시 주소 계산
  count = M
  while count > 0:
    if table[i] == None:
      break
    i = (i + 1) % M   # 다음 위치 조사
    count -= 1

  if count > 0:
    table[i] = key

# 탐색 연산
def search(key) :
    i = hashFn(key)
    count = M
    while count>0 :
        if table[i] == None :
            return None
        if table[i] == key :
            return table[i]
        i = (i + 1) % M
        count -= 1

    return None

# 삭제 연산
def delete(key) :
    i = hashFn(key)
    count = M
    while count>0 :
        if table[i] == key :
            table[i] = -1
            return
        if table[i] == None :
            return
        i = (i + 1) % M
        count -= 1
```

```python
data = [1, 23, 456, 7890]

# 삽입
for i in data:
  insert(i)
print(table)
```

```python
# 탐색
for i in data:
  print(search(i))
```

```python
# 삭제
delete(23)
print(search(23))
```
