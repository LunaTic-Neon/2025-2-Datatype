```python
class TNode:
  def __init__(self,elem,left,right):
    self.data = elem
    self.left = left
    self.right = right

  def isLeaf(self):
    return self.left is None and self.right is None

# 이진트리 전위순회
def preorder(n):
  if n is not None:
    print(n.data, end='')
    preorder(n.left)
    preorder(n.right)

# 이진트리 중위순회
def inorder(n):
  if n is not None:
    preorder(n.left)
    print(n.data, end='')
    preorder(n.right)

# 이진트리 후위순회
def postorder(n):
  if n is not None:
    postorder(n.left)
    postorder(n.right)
    print(n.data, end='')

# 노드 수 계산
def count_node(n) :
    if n is None :
        return 0
    else :
        return 1 + count_node(n.left) + count_node(n.right)

# 단말노드 수 계산
def count_leaf(n) :
    if n is None : return 0
    elif n.isLeaf() : return 1
    else : return count_leaf(n.left) + count_leaf(n.right)

# 트리 높이 계산
def calc_height(n) :
    if n is None : return 0
    hLeft = calc_height(n.left)
    hRight = calc_height(n.right)
    if (hLeft > hRight) : return hLeft + 1
    else: return hRight + 1
```

```python
# 노드 만들기
d = TNode('D',None,None)
e = TNode('E',None,None)
b = TNode('B',d,e)

f = TNode('F',None,None)
g = TNode('G',None,None)
c = TNode('C',f,g)

root = TNode('A',b,c)

#      A
#    B   C
#  D  E F  G

print("중위순회 : ", end='')
inorder(root)
print()

print("후위순회 : ", end='')
postorder(root)
print()

print("전위순회 : ", end='')
preorder(root)
print()
```

```python
# 노드 수 계산
print("노드 수 :", count_node(root))
```

```python
# 단말노드 수 계산
print("단말노드 수 :", count_leaf(root))
```

```python
# 트리 높이 계산
print("트리 높이 :", calc_height(root))
```
